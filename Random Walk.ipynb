```python
n = 4 

def Clash(queen_pos):
    board = []
    
    for i in range(n):
        board.append((queen_pos[i], i))
#     print(board)

    count = 0
    
    for i in range(n-1):
        queen1 = board[i]
        
        for j in range(i+1, n):
            queen2 = board[j]
            
            if(abs(queen1[0] - queen2[0]) == abs(queen1[1] - queen2[1])):
                count += 1
            
            if queen1[0] == queen2[0]:
                count += 1
#     print(count)
    return count


queen_pos = [1, 2, 3, 1]
Clash(queen_pos)
```

```python
# Finding the successive placement for Queens

positions = []

def Successor(queen_pos):
    positions.clear()
    lst = []
    
    for i in range(n):
        for j in range(n):
            if queen_pos[i] != j :
                lst = queen_pos.copy()
                lst[i] = j
                positions.append(lst)
    
    for k in positions:
        print(k, Clash(k))
    
    return positions
```
