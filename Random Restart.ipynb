```python
n = 4 

def Clash(queen_pos):
    board = []
    
    for i in range(n):
        board.append((queen_pos[i], i))
#     print(board)

    count = 0
    
    for i in range(n-1):
        queen1 = board[i]
        
        for j in range(i+1, n):
            queen2 = board[j]
            
            if(abs(queen1[0] - queen2[0]) == abs(queen1[1] - queen2[1])):
                count += 1
            
            if queen1[0] == queen2[0]:
                count += 1
#     print(count)
    return count


queen_pos = [1, 2, 3, 1]
Clash(queen_pos)
```

```python
# Finding the successive placement for Queens

positions = []

def Successor(queen_pos):
    positions.clear()
    lst = []
    
    for i in range(n):
        for j in range(n):
            if queen_pos[i] != j :
                lst = queen_pos.copy()
                lst[i] = j
                positions.append(lst)
    
    for k in positions:
        print(k, Clash(k))
    
    return positions
```

```python
# Hill climbing with random restart

import random

num = int(input("Enter total no. of queens : "))
print("\n")

pos = []
for i in range(num):
    print("Enter row number of queen ", i+1)
    row = int(input())
    
    pos.append(row)
print("\n")
clash_count = Clash(pos)
# print(clash_count)

max_h = 99

while(1):
    max_h = 99
    clash_count = Clash(pos)
    while(max_h > clash_count):
        positions = []

        positions = Successor(pos)

        min_h = 99
        minH_lst = []

        for lst in positions:
            minH_val = Clash(lst)

            if minH_val < min_h:
                min_h = minH_val
                minH_lst = lst.copy()

        max_h = clash_count
        clash_count = min_h

        pos = minH_lst.copy()

        print("Best Position : ", pos, clash_count)
        print("\n")
        if(clash_count == 0):
            break
    
    if clash_count == 0:
        print("\n")
        print("Solution found : ")
        print(pos, clash_count)
        break

    else:
#         print("Solution to problem is : ")
        pos = []
        pos.append(random.randint(0,3))
        pos.append(random.randint(0,3))    
        pos.append(random.randint(0,3))    
        pos.append(random.randint(0,3))    
        print("Randomly Restart with position : ", pos)
```
